## ğŸ”¬ Research Overview

### Main Research Question

> **Primary:** How does implicit cooperation, enabled by MARL, improve the management of DERs to maximize their energy use while ensuring the balance between supply and demand in LEMs?

> **Complementary:** How do different market mechanism designs and incentive structures influence agent participation strategies, energy allocation decisions and grid operations in decentralized local energy markets to maintain grid balance while maximizing economic efficiency?

### Research Objectives

- **Validate implicit cooperation effectiveness** in decentralized energy markets
- **Compare market mechanisms** for optimal pricing and allocation
- **Analyze agent heterogeneity** and market power dynamics
- **Study DSO intervention strategies** for grid stability
- **Examine grid topology effects** on market efficiency
- **Investigate battery coordination** for enhanced flexibility

### Key Innovation

This research introduces **implicit cooperation** in decentralized energy markets through MARL, enabling agents to coordinate without explicit communication while maintaining system efficiency and grid stability.


## ğŸ“ Training Methodology

Each case study uses **three training paradigms** for comprehensive validation:

### 1. CTCE (Centralized Training, Centralized Execution)
- **Purpose:** Baseline for optimal coordination
- **Characteristics:** Centralized critic with shared information
- **Use Case:** Theoretical upper bound performance

### 2. CTDE (Centralized Training, Decentralized Execution)
- **Purpose:** Main practical approach
- **Characteristics:** Centralized training, decentralized deployment
- **Use Case:** Real-world implementation feasibility

### 3. DTDE (Decentralized Training, Decentralized Execution)
- **Purpose:** Test true implicit coordination
- **Characteristics:** Fully decentralized approach
- **Use Case:** Core research validation

### Training Parameters

| Parameter | Value | Description |
|-----------|-------|-------------|
| Training Episodes | 1000 | Sufficient for cooperation learning |
| Evaluation Episodes | 100 | For performance assessment |
| Learning Rate | 0.001 | Standard RL learning rate |
| Batch Size | 64 | Training batch size |
| Convergence Threshold | 0.8 | Cooperation quality threshold |
